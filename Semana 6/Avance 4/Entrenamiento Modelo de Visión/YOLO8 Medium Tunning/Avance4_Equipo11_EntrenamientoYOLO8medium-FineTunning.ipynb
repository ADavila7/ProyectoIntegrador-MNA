### **Modelo 2. YOLO V8 medium**

```python
# Load the YOLOv8 model
model5 = YOLO('yolov8m.pt')
```

```python
# Define hyperparameters
learning_rate = 0.01
final_learning_rate = 0.001  # Tasa de aprendizaje final

# Train the model with custom learning rates and other parameters
model5.train(
    data=Modelo_5,
    epochs=100,
    imgsz=640,
    lr0=learning_rate,
    lrf=final_learning_rate,
    momentum=0.937,
    weight_decay=0.0005
)
```
